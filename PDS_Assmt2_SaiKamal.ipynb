```python
#importing all the required libraries
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from datetime import datetime
```

```python
#Reading the dataset
data_Cars = pd.read_csv("train.csv")
```

```python
data_Cars.head()
```

```python
#checking for null values in the data frame
null_Car_data = data_Cars.isnull().sum()
print(null_Car_data)
```

```python
#dropping New_Price column because it has more NULL values compared to other numeric columns
data_Cars = data_Cars.drop('New_Price', axis=1)
```

```python
data_Cars.head()
```

```python
#removing characters after interger values in the above columns
data_Cars['Mileage']=data_Cars['Mileage'].str.split(' ').str[0]
data_Cars['Engine']=data_Cars['Engine'].str.split(' ').str[0]
data_Cars['Power']=data_Cars['Power'].str.split(' ').str[0]
data_Cars['Seats']=data_Cars['Seats'].fillna(0)
```

```python
# converting string columns values to float
data_Cars['Mileage'] = data_Cars['Mileage']. astype(float)
data_Cars['Engine'] = data_Cars['Engine']. astype(float)
data_Cars['Power'] = data_Cars['Power']. astype(float)
data_Cars['Seats'] = data_Cars['Seats'] . astype(float)
```

```python
#replacing the NULL values with mean

numeric_Values = ['Mileage', 'Engine', 'Power', 'Price','Seats']
imputer = SimpleImputer(strategy='mean')
data_Cars[numeric_Values] = imputer.fit_transform(data_Cars[numeric_Values])
```

```python
#checking for null values in the data frame after dropping and adding mean values in the place of NULL
data_Cars.isnull().sum()
```
